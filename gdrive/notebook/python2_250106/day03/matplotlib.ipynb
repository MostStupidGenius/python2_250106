## 그래프의 종류
- 선 그래프
    - 점과 점을 이어 값의 변화를 관찰하는 데 용이
- 막대 그래프
    - 범주별 값의 크기의 비교나 빈도를 효과적으로 나타낼 수 있다.
- 산점도
    - 두 변수의 관계를 좌표로 찍어 그 데이터가 흩어진 정도를 한눈에 파악할 수 있다.
    - 두 변수의 관계를 시각화하는 데 주로 쓰인다.
- 파이 차트
    - 원형의 그래프로, 각각의 범주가 총 데이터량에서 차지하는 비율을 나타내는 데 쓰인다.
    - 예시) 투표율, 참여율,... 뉴스에서 많이 나온다.
- 히스토그램
    - 막대 그래프와 유사하지만 그 데이터의 종류가 범주형이 아닌 연속형 데이터라는 큰 차이점이 있다.
    - 때문에 연속형 데이터의 범위를 조정하면 막대의 너비를 조절할 수 있고, 그 촘촘한 정도가 달라질 수 있다.

### 선 그래프
- Sin 함수 그래프 그리기


````python
# 필요한 라이브러리 불러오기
import matplotlib.pyplot as plt
import numpy as np
amount = 100 # @param {"type":"slider","min":10,"max":500,"step":10}
# x축 데이터 생성 (0부터 10까지 100개(기본)의 균등한 간격의 점)
x = np.linspace(0, 10, amount)
# y축 데이터 생성 (sin 함수 적용)
y = np.sin(x)

# print(x)
# print("==============")
# print(y)
````

````python
# 그래프 그리기
# x데이터의 개수와 y데이터의 개수가 일치해야 한다.
plt.plot(x, y)

# 그래프 제목 설정
plt.title('Sin 함수 그래프')

# x축 레이블 설정
plt.xlabel('X 축')

# y축 레이블 설정
plt.ylabel('Y 축')

# 그래프 표시
plt.show()
````

### 막대 그래프
- 범주형 데이터들의 수량이나 빈도 등을 직관적으로 비교할 수 있는 그래프다.
- 각 범주는 높이나 길이가 데이터 값에 비례하는 막대로 표현된다.
- 세로 막대, 가로 막대, 누적 막대 등 다양한 형태로 표현이 가능하다.

- 이러한 막대 그래프로 그래프를 표현하기 위해서는 데이터프레임이라는 형태로 데이터를 정제해야 한다.
- 한 열(컬럼)의 데이터를 대표하는 단어(헤더)와 그 데이터들의 나열로 이루어진 열들을 표 형태로 만들어둔 것을 데이터 프레임이라 부른다.


````python
# 과일의 종류와 그 개수를 각각의 리스트에 담아서
# dict 자료형에 담아
# 데이터프레임으로 변환하여 막대 그래프의 데이터로 활용한다.
# 데이터프레임으로 변환하기 위한 라이브러리 판다스
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
data = {
    '과일': ['사과', '배', '감'],
    '판매량': [30, 50, 42]
}
# 데이터프레임으로 변환
df = pd.DataFrame(data)

# .bar() 막대 그래프 그리는 함수
plt.bar(df['과일'], df['판매량'])

plt.title("과일 판매량")
plt.xlabel("과일")
plt.ylabel("판매량")


# 그래프 표현
plt.show()
````

### 산점도(scatter)
- 두 변수(x, y)에 의해 찍힌 점(좌표)의 분포를 직관적으로 확인하기 용이한 그래프다.
- seaborn 라이브러리를 이용하여 산점도를 그릴 수 있다.
```python
import seaborn as sns
sns.scatterplot(x="x값으로쓸data의변수명", y="y값으로쓸data의변수명", data = data)
```

````python
# 산점도를 그리기 위한 seaborn 라이브러리
import seaborn as sns

data = sns.load_dataset("tips")
print(data)
````

````python
# 데이터를 입력하여 산점도 그리기
sns.scatterplot(x="total_bill", y="tip", data = data)

plt.title("계산서와 팁의 관계")
plt.show()
````

#### Plotly 인터렉티브 시각화
- Plotly 라이브러리를 활용해 인터랙티브(대화형)한 그래프를 만들 수 있다.
- 인터랙티브: 데이터를 필터링하여 원하는 데이터를 추출 후 시각화에 활용할 수 있다.

````python
# Plotly 라이브러리 설치
!pip install plotly
````

````python
# Plotly Express 모듈 임포트
import plotly.express as px

# Gapminder 데이터셋에서 2007년 유럽 국가 데이터만 선택
df = px.data.gapminder().query("continent=='Europe' & year==2007")
print(df)
````

````python
# 산점도 생성
fig = px.scatter(
    df, # 시각화할 데이터프레임
    x="gdpPercap",  # x축: 1인당 GDP
    y="lifeExp",    # y축: 기대 수명
    size="pop",     # 점크기: 인구에 비례
    color="country", # 색은 나라별로 다르게 칠한다.
    hover_name="country", # 마우스를 올리면 표시할 정보: 국가명
    log_x=True,     # x축 로그 스케일 적용
    size_max=60     # 최대 점 크기 설정
)
# 그래프 표시
fig.show()
````

### 히스토그램


````python
import numpy as np
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight)

# 넘파이의 히스토그램
counts, bins = np.histogram(weight)
# 외곽선 그리기
plt.stairs(counts, bins)
# y축의 눈금을 조정
plt.yticks(np.arange(0, 5, 1)) # 0부터 5미만까지 1만큼의 간격으로 눈금을 그린다.


plt.show()
````
